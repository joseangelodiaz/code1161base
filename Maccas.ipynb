# Calories

```python
item_calorie_graph = burgers.groupby("Item")["Calories"].sum()
```

```python
item_calorie_graph
```

```python
calorie_graph = item_calorie_graph.sort_values(ascending=False)
```

```python
calorie_graph.plot(kind="barh", fontsize="15", facecolor= "yellow")
plt.title("Calorie Intake of McDonald's Beef Burgers", fontsize=20)
plt.xlabel("Calories", fontsize=20)
plt.ylabel("Burger Name", fontsize=20)
```

```python
item_carbohydrates_graph = burgers.groupby("Item")["Carbohydrates"].sum()
```

# Carbohydrates

```python
item_carbohydrates_graph
```

```python
carb_graph = item_carbohydrates_graph.plot(kind="bar", fontsize="12", facecolor="Green")
plt.title("Amount of Carbohydrates in McDonald's Beef Burgers", fontsize=20)
plt.xlabel("Burger Names", fontsize=20)
plt.ylabel("Carbohydrates (g)", fontsize=20)
```

# Protein

```python
item_protein_graph = burgers.groupby("Item")["Protein"].sum()
```

```python
item_protein_graph
```

```python
from pandas.tools.plotting import table
#code for pie chart
plt.figure(figsize=(16,8))
ax1 = plt.subplot(121, aspect="equal")
item_protein_graph.plot(kind='pie', title="Protein(g) in McDonald's Beef Burgers", ax=ax1, autopct="%1.1f%%", startangle=90, shadow=True, labels=burgers["Item"], legend=False, fontsize=16)
#table code for proteins table
ax2 = plt.subplot(122)
plt.axis("off")
tb1 = table(ax2, item_protein_graph, loc="right")
tb1.auto_set_font_size(False)
tb1.set_fontsize(14)
```
